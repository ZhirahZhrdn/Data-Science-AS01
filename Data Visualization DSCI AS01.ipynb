```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("students_mental_health_survey.csv")

average_stress_level_per_course = df.groupby("Course")["Stress_Level"].mean()

courses = average_stress_level_per_course.index.to_list()

average_stress_levels = average_stress_level_per_course.to_list()

# Create a bar chart
plt.bar(courses, average_stress_levels)
plt.xlabel("Course")
plt.ylabel("Average Stress Level")
plt.xticks(rotation=45)

plt.title("Relationship Between Course and Stress Level")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("students_mental_health_survey.csv")

# Group the data by age and calculate the sum stress level for each age group
df_grouped = df.groupby(['Relationship_Status'])['Financial_Stress'].sum()
age = df_grouped.index.to_list()
stress_level = df_grouped.to_list()

# Create a bar chart
plt.bar(age, stress_level)

plt.xlabel('Relationship Status')
plt.ylabel('Average Financial Level')
plt.title('Relationship Between Relationship Status and Financial Stress')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("students_mental_health_survey.csv")

# Group the data by age and calculate the sum stress level for each age group
df_grouped = df.groupby(['CGPA'])['Anxiety_Score'].sum()
cgpa = df_grouped.index.to_list()
anxiety = df_grouped.to_list()

# Create a bar chart
plt.scatter(cgpa, anxiety)

plt.xlabel('CGPA')
plt.ylabel('Anxiety Score')
plt.title('Relationship Between CGPA and Anxiety Score')

plt.show()
```
